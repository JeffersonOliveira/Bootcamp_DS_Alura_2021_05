```python
import pandas as pd
```

```python
data = pd.read_csv("/content/A003954189_28_143_208.csv", encoding='latin-1', skiprows=3, sep=";",
            skipfooter=12, engine="python", thousands=".", decimal=",")
data.head()
```

```python
data.tail()
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
data['2008/Jan']
```

```python
data.sort_values('2008/Ago', ascending=False).plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = data.sort_values('2008/Ago', ascending=False).plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor gasto em Agosto de 2008 por Unidade da Federação", fontsize="20" )

# axs1[1].xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.3f}"))

plt.show()
```
